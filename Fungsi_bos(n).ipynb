```python
import numpy as np
from numpy import cos, exp
import random
import matplotlib.pyplot as plt

def bos(n): #mendefinisikan fungsi 'bos(n)''
    for i in range(n):
        plt.xlim(0,1) #ukuran plot pada sumbu x = 1
        plt.ylim(0,2) #ukuran plot pada sumbu y = 2
        
        titik_x = []
        titik_y = []
        int_x = []
        int_y = []
        
        nilai_a = random.randint(1,9) #merandom nilai a : 1<=a<10
        a = str(nilai_a) #mengubah nilai random a menjadi string
        nilai_b = random.randint(1,9) #merandom nilai b : 1<=b<10
        b = str(nilai_b) #mengubah nilai random b menjadi string

        fungsi = "exp(-({0}*x**2))*abs(cos({1}*x)) + 0.5".format(a,b) #fungsi pada soal
        def f(x):
            return eval(fungsi)
        print('Untuk nilai a = {0} dan b = {1}'.format(a,b))
        print('y = ',fungsi)

        for j in np.linspace(0,1,100):
            titik_x.append(j) #menginput titik j ke dalam list kosong 'titik_x'
            titik_y.append(f(j)) #menginput nilai fungsi j ke dalam list kosong 'titik_y'
        for z in np.linspace(0,1,100): #gunakan Composite Simpson
            h = z/100 #h = (b-a)/n, b = z (z adalah entri list dari interval 0 sampai 1), n = 100 (banyaknya entri list)
            sum1 = 0
            sum2 = 0
            for j in range(1, int(100/2)):
                sum1 += f(2*j*h)
                sum2 += f((2*j-1)*h)
            sum2 += f(1-h)
            int_x.append(z)
            int_y.append((h/3)*(f(0)+2*sum1+4*sum2+f(z)))

        result = ((1/100)/3)*(f(0)+2*sum1+4*sum2+f(1))

        plt.plot(titik_x,titik_y, color='r') #plot titik x dan f(x)
        plt.plot(int_x,int_y, color='b') #plot integral dari f(x)
        plt.legend(['f(x)','integral dari f(x)']) #keterangan mengenai kedua plot
        plt.grid(True) #membuat grid koordinat
        plt.show()
        print('Luas daerah = {0:.5f} hektar'.format(result))
        print('========================================================')
        print(' ')
```
