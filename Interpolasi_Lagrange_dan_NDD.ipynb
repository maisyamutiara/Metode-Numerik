Program ini adalah program untuk menghitung nilai interpolasi dengan mengaproksimasi nilai x menggunakan beberapa pasangan titik x dan titik y yang diketahui, sehingga didapatkan nilai interpolasi y. Program ini dapat mencari nilai interpolasi y dengan 3 metode yaitu Lagrange, Newton Divided Difference Forward, dan Newton Divided Difference Backward.

1. Metode Interpolasi Lagrange

Interpolasi Lagrange didefinisikan sebagai berikut :

$$ P(x) = f(x_0) L_{n, 0}(x) + f(x_1) L_{n, 1}(x) + \cdots + f(x_n) L_{n, n}(x) = \sum_{k = 0}^n f(x_k) L_{n, k}(x) $$

dimana untuk setiap $$ k = 0, 1,... , n $$
$$ L_{n, k}(x) = \prod_{i = 0, i \neq k}^n \cfrac{x - x_i}{x_k - x_i} $$

2. Newton Divided Difference (NDD)

NDD memiliki bentuk sebagai berikut :

$ P_n(x) = f[x_0] + \sum_{k = 1}^n f[x_0, x_1, \cdots, x_k] (x - x_0) (x - x_1) \cdots (x - x_{k - 1}) $$

dimana $$ f[x_k] = f(x_k) $$ dan 
$$ f[x_i, x_{i + 1}, \cdots, x_{i + k}] = \cfrac{f[x_{i + 1}, x_{i + 2}, \cdots, x_{i + k}] - f[x_i, x_{i + 1}, \cdots, x_{i + k - 1}]}{x_{i + k} - x_i} $$

Ada 2 cara pembentukan polinomal interpolasi dengan DD, yaitu metode Forward dan Backward.

   1. Forward DD menggunakan baris paling atas dari setiap kolom pada tabel DD.
   
   $$ P_n(x) = f[x_0] + \sum_{k = 1}^n f[x_0, x_1, \cdots, x_k] (x - x_0) (x - x_1) \cdots (x - x_{k - 1}) $$
    
   2. Backward DD menggunakan baris paling akhir dari setiap kolom pada tabel DD.
   
   $$ P_n(x) = f[x_n] + \sum_{k = 1}^n f[x_n, x_{n - 1}, \cdots, x_{n - k}] (x - x_n) (x - x_{n - 1}) \cdots (x - x_{n - k + 1}) $$

Algoritmanya adalah
1. Menginput list berisi titik x, titik y, dan titik x yang akan diaproksimasi
2. Membuat for loop untuk menjalankan program
3. Menggunakan if conditional untuk memilih metode Lagrange atau NDD
4. Jika diinput 1 untuk variabel "metode" maka akan digunkan metode interpolasi Lagrange, program akan mencari nilai interpolasi dengan rumus Lagrange dan mencari rumus persamaan fungsi polinomnya.
5. Jika diinput 2 untuk variabel "metode" maka akan digunakan metode interpolasi NDD, program akan menjalankan tabel segitiga Divided Difference.
6. Membuat for loop kedua untuk menjalankan if conditional metode interpolasi Forward NDD atau Backward NDD.
7. Jika diinput 1 untuk variabel "DDmetode" maka akan digunkan metode interpolasi Forward NDD, program akan mencari nilai interpolasi dengan rumus Forward NDD dan mencari rumus persamaan fungsi polinomnya.
8. Jika diinput 2 untuk variabel "DDmetode" maka akan digunkan metode interpolasi Backward NDD, program akan mencari nilai interpolasi dengan rumus Backward NDD dan mencari rumus persamaan fungsi polinomnya.
9. Jika dipilih angka lain untuk variabel "DDmetode" maka program akan keluar dari looping
10. Membuat if conditional untuk memilih akan ditampilkan tabel divided difference atau tidak dengan menginput string pada variabel "table".
11. Jika diinput "Y" atau "y" untuk variabel "table" maka program akan menampilkan tabel Divided Difference.
12. Jika diinput "N" atau "n" untuk variabel "table" maka program tidak akan menampilkan tabel dan keluar dari looping.
13. Jika diinput string lainnya untuk variabel "table" maka program akan keluar dari looping.
14. Membuat if conditional untuk memilih menampilkan visualisasi berupa tabel result atau plot.
15. Jika diinput 1 untuk variabel "show" maka program akan menampilkan tabel result yang berisikan titik x keseluruhan dan titik y keseluruhan.
16. Jika diinput 2 untuk variabel "show" maka program akan menampilkan plot yang terdiri dari garis plot fungsi, garis plot dari yang titik diinput, dan pasangan titik x yang diaproksimasi dan titik y hasil interpolasi.
17. Jika diinput angka lainnya untuk variabel "show" maka program akan keluar dari looping.

```python
import sympy
from tabulate import tabulate
from sympy import simplify
import matplotlib.pyplot as plt
import numpy as np

titik_x = eval(input('Masukkan list titik-titik berurutan : ')) #nilai x
titik_y = eval(input('Masukkan list nilai fungsi di titik-titik tersebut : ')) #nilai y
eval_x = eval(input('Masukkan nilai x yang akan diaproksimasi nilai fungsinya : ')) #nilai x yang dicari
metode = eval(input('Pilih metode yang diinginkan (1 = Lagrange, 2 = NDD) : ')) #pilih metode
g = 999999999999999999999999999999999999999999 #inisialisasi batas loop

for coba in range(g): #for loop
    if metode == 1 :
        def LagrangePol(x, x_points, y_points) :
            pol = 0 #untuk inisialisasi nilai polinom diset menjadi 0
            n = len(x_points) #n adalah panjang list x atau banyaknya titik
            for k in range(n) : #iterasi sebanyak panjang list x
                L = 1 #untuk inisialisasi nilai L(n,k) diset menjadi 1
                for i in range(n) : #iterasi untuk L
                    if i != k :
                        L *= ((x-x_points[i])/(x_points[k]-x_points[i])) #menghitung nilai L(n,k)
                pol += y_points[k] * L #menghitung nilai aproksimasi dengan polinom
            return pol
        
        #mencari bentuk polinom persamaan fungsi Lagrange
        def persamaanpolinom(x_points, y_points) :
            x = sympy.symbols('x') #simbol untuk suku x
            if  len(x_points)!= len(y_points) : #banyak titik x harus sama dengan banyak titik y
                return 1
            pol = 0
            for k in range(len(x_points)) :
                L = 1
                for j in range(len(x_points)) :
                    if j != k :
                        L = L*((x-x_points[j]) /(x_points[k]-x_points[j]))
                pol += L*y_points[k]
            return pol
        polinom = persamaanpolinom(titik_x,titik_y)
        Fx = str(simplify(polinom)) #fungsi polinom lagrange
        print('Lagrange Polynomial = :',Fx)
        y_aproks = LagrangePol(eval_x, titik_x, titik_y) #nilai y hasil aproksimasi titik x
        print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x, y_aproks))

    elif metode == 2 :
        def DDTableGenerator(x_points, y_points) : #buat fungsi untuk membuat tabel DD
            DDTable = [y_points] #kolom-kolom pada tabel, kolom pertama berisi titik y
            for column in range(1, len(y_points)) :
                DDcolumn = [] #list kosong yang akan diisi dari setiap kolom
                for row in range(len(DDTable[-1])-1) : #mulai mengisi kolom tabel
                    DD = (DDTable[-1][row+1]-DDTable[-1][row])\
                    /(x_points[column+row]-x_points[row])
                    DDcolumn.append(DD)
                DDTable.append(DDcolumn) #tambahkan kolom yang telah diisi ke tabel
            return DDTable
        
        DDmetode = eval(input('Pilih Forward atau Backward (1 = Forward, 2 = Backward) : '))
        for run in range(g):
            if DDmetode == 1 :
                def ForwardDD(x, x_points, y_points) :
                    DDTable = DDTableGenerator(x_points, y_points)
                    pol = DDTable[0][0] #nilai dari polinom. Inisialisasi : suku pertama polinom
                    mult_term = 1 #variabel untuk menyimpan nilai dari (x-x0)(x-x1)...(x-x(n-1))
                    for k in range(1, len(DDTable)) :
                        mult_term *= (x-x_points[k-1]) #menghitung (x-x0)(x-x1)...(x-x(n-1))
                        pol += DDTable[k][0] * mult_term #menghitung nilai interpolasi
                    return pol
                
                #mencari bentuk polinom persamaan fungsi Forward DD
                def persamaanpolinom(x_points, y_points):
                    x = sympy.symbols('x') #simbol untuk suku x
                    DDTable = DDTableGenerator(x_points, y_points)
                    pol = DDTable[0][0]
                    mult_term = 1
                    for k in range(1, len(DDTable)):
                        mult_term = mult_term*(x-x_points[k-1])
                        pol = pol + DDTable[k][0]*mult_term
                    return pol
                polinom = persamaanpolinom(titik_x,titik_y)
                Fx = str(simplify(polinom)) #fungsi polinom Forward DD
                print('Forward NDD Polynomial = :',Fx)
                y_aproks = ForwardDD(eval_x, titik_x, titik_y) #nilai y hasil aproksimasi titik x
                print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x, y_aproks))
        
            elif DDmetode == 2 :        
                def BackwardDD(x, x_points,y_points) :
                    DDTable = DDTableGenerator(x_points, y_points)
                    pol = DDTable[0][-1] #nilai dari polinom. Inisialisasi : suku pertama polinom
                    mult_term = 1 #variabel untuk menyimpan nilai dari (x-xn)(x-x(n-1))...(x-x1)
                    for k in range(1, len(DDTable)) :
                        mult_term *= (x-x_points[-k]) #menghitung (x-xn)(x-x(n-1))...(x-x1)
                        pol += DDTable[k][-1] * mult_term #menghitung nilai interpolasi
                    return pol
                
                #mencari bentuk polinom persamaan fungsi Backward DD
                def persamaanpolinomial(x,x_points,y_points) :
                    DDTable = DDTableGenerator(x_points,y_points)
                    pol = DDTable[0][-1]
                    mult_term = 1
                    for k in range(1,len(DDTable)) :
                        mult_term *= (x-x_points[-k])
                        pol += DDTable[k][-1] * mult_term
                    return pol
                polinom = persamaanpolinom(titik_x,titik_y)
                Fx = str(simplify(polinom)) #fungsi polinom Backward DD
                print('Backward NDD Polynomial = :',Fx)
                y_aproks = BackwardDD(eval_x, titik_x, titik_y) #nilai y hasil aproksimasi titik x
                print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x, y_aproks))
        
            else :
                print("Tidak ada di pilihan angka")
                break #keluar dari looping
        
            #membuat tabel pyramid Divided Difference
            table = str(input('Menampilkan tabel Divided Difference? (Y/N) : '))
            if table in ["y","Y"] : #y dapat berupa kapital atau huruf kecil
                def tabelDD(DDTable, x_point):
                    table = [] #list kosong yang akan diinput segitiga DD
                    for i in range(len(DDTable)):
                        pyramid = [] #list kosong yang akan diinput nilai DD
                        for j in range(len(DDTable[i])):
                            pyramid.append(DDTable[j][i]) #memasukkan nilai DD ke list kosong "pyramid"
                        table.append(pyramid) #memasukkan nilai segitiga DD ke list kosong "table"
                    for x in range(len(table)):
                        table[x].insert(0, x_point[x])
                    return print(tabulate(table, headers=["x", "y","DD1", "DD2", "DD3", "DD4", "DD5", "DD6", "DD7", "DD8"], tablefmt='orgtbl'))

                fixtabel = tabelDD(DDTableGenerator(titik_x,titik_y),titik_x)
                print(fixtabel)
                break
            elif table in ["n","N"] : #n dapat berupa kapital atau huruf kecil
                None
                break
            else :
                print("Tidak ada di pilihan")
                break
        
    else :
        print("Tidak ada di pilihan angka")
        break
    
    show = eval(input('Tampilkan visualisasi (1 = Tabel, 2 = Plot) : '))
    
    #membuat tabel result titik x total dan titik y (f(x)) total
    if show == 1 :
        t = np.append(titik_x, eval_x) #menginput nilai "titik_x" dan "eval_x" ke numpy kosong "t"
        s = np.append(titik_y, y_aproks) #menginput nilai "titik_y" dan "y_aproks" ke numpy kosong "s"
        entri_x = np.sort(t) #mengurutkan angka di list "t"
        entri_y = np.sort(s) #mengurutkan angka di list "s"
        print("RESULT")
        print(tabulate({"x":entri_x,"y":entri_y},headers='keys',tablefmt='github'))
        break
        
    #membuat plot grafik input dan polinomial interpolasi
    elif show == 2 :
        def p(x):
            return eval(Fx)
        x2 = np.linspace(titik_x[0],titik_x[-1]) 
        plt.plot(x2,p(x2)) #plot fungsi polinom
        plt.plot(eval_x,y_aproks,'r-o') #plot titik x dan y yang diinterpolasi
        plt.plot(titik_x, titik_y,'g-') #plot titik x dan y yang diinput
        plt.legend(['polinom interpolasi','titik yang diinterpolasikan','titik input'])
        plt.xlabel('Nilai x')
        plt.ylabel('Nilai y')
        plt.title('Grafik Input dan Polinomial Interpolasi')
        plt.show()
        break
    else :
        print("Tidak ada di pilihan angka")
        break
```
